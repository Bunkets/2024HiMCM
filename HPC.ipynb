```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
file_path = "HPC.xlsx"  # Replace with the actual path to your file
data = pd.read_excel(file_path)

# Group by "Estimate year" and average "Value (TWh)"
averaged_data = data.groupby("Estimate year")["Value (TWh)"].mean().reset_index()

# Additional data points as dictionaries
added_data = {
    2008: 0.6715792679999999,
    2009: 0.8210551337999998,
    2010: 0.9937807404000002,
    2011: 1.3872098166,
    2012: 1.7089403075999998,
    2013: 1.8134776799999999,
    2014: 1.8777081461999998,
    2015: 1.8088013292000003,
    2016: 1.9658159196,
    2017: 3.3212527052,
    2018: 3.636255633,
    2019: 3.0596863175999998,
    2020: 2.9317345602000002,
    2021: 2.8653569304,
    2022: 2.9884500414,
    2023: 3.159339333,
    2024: 3.5152506432,
    2025: 3.693060,
    2026: 3.870869,
    2027: 4.048678,
    2028: 4.226488,
    2029: 4.404297,
    2030: 4.582106
}

# Create a DataFrame for added data
added_df = pd.DataFrame(list(added_data.items()), columns=["Year", "Added Value"])

# Merge original data with added data
merged_data = pd.merge(averaged_data, added_df, left_on="Estimate year", right_on="Year", how="right")

# Calculate the ratio (Original Value / Added Value)
merged_data["Ratio"] = merged_data["Value (TWh)"] / merged_data["Added Value"]

# Plot the ratio
plt.figure(figsize=(10, 6))
plt.scatter(merged_data["Year"], merged_data["Ratio"], facecolors='none', edgecolors='purple', s=80, label="Ratio")

# Set axis bounds and tick increments
plt.xlim(2006, 2032)  # Add space before and after the edge points
plt.xticks(range(2006, 2033, 2))  # Increment the x-axis by 2 years

# Add title and labels
plt.title("Year vs Ratio of Global Dataset to TOP500", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Ratio (Global Dataset / TOP500)", fontsize=12)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Save and display the plot
plt.savefig("HPC_ratio_graph.png")
plt.show()

merged_data
```
