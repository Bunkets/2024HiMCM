```python
import matplotlib.pyplot as plt

# Data
scenarios = ["Current Energy Mix", "50% Renewable Energy Mix"]
emissions = [160147474, 109348242]
colors = ["lightcoral", "lightgreen"]

# Plot
plt.figure(figsize=(8, 6))
plt.bar(scenarios, emissions, color=colors)

# Title and Labels
plt.title("2024 HPC Carbon Emissions: Comparison", fontsize=14)
plt.ylabel("Metric Tons CO2", fontsize=12)
plt.xlabel("Scenario", fontsize=12)

# Format y-axis for better readability
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Display the plot
plt.tight_layout()
plt.savefig("5B.png")
plt.show()
```
