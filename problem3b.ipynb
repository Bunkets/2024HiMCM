```python
import matplotlib.pyplot as plt

# Data for the bar graph
scenarios = ["Best-Case Scenario", "Worst-Case Scenario"]
emissions = [502501952, 689548131]  # Emissions in metric tons of CO2

# Create the bar graph
plt.figure(figsize=(8, 6))
plt.bar(scenarios, emissions, color=['lightgreen', 'lightcoral'], alpha=0.8)  # Softer colors with transparency

# Add title and labels
plt.title("2030 Carbon Emission Bounds", fontsize=16)
plt.ylabel("Carbon Emissions (Metric Tons of CO2)", fontsize=12)
plt.xlabel("Scenario", fontsize=12)

# Format the y-axis with commas for large numbers
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.ticklabel_format(style='plain', axis='y')  # Prevent scientific notation

# Annotate bar values
for i, value in enumerate(emissions):
    plt.text(i, value + 20000, f"{value:,.2f}", ha='center', fontsize=10)

# Display the plot
plt.tight_layout()
plt.savefig("emission_bounds")
plt.show()
```
