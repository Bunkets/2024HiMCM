```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'lst_us_data_centers_20241115_070225.csv'  # Update the file path as needed
data = pd.read_csv(file_path)

# Filter rows where both "TOT_DATACENTER_SPACE_SQFT" and "TOT_POWER_MW" have values
filtered_data = data.dropna(subset=["TOT_DATACENTER_SPACE_SQFT", "TOT_POWER_MW"])

# Convert columns to numeric
filtered_data["TOT_DATACENTER_SPACE_SQFT"] = pd.to_numeric(filtered_data["TOT_DATACENTER_SPACE_SQFT"], errors='coerce')
filtered_data["TOT_POWER_MW"] = pd.to_numeric(filtered_data["TOT_POWER_MW"], errors='coerce')

# Drop rows with invalid numeric data
filtered_data = filtered_data.dropna(subset=["TOT_DATACENTER_SPACE_SQFT", "TOT_POWER_MW"])

# Apply Z-score method to remove outliers
z_scores = np.abs(zscore(filtered_data[["TOT_POWER_MW", "TOT_DATACENTER_SPACE_SQFT"]]))
filtered_data_no_outliers = filtered_data[(z_scores < 2).all(axis=1)]

# Define X (independent variable) and y (dependent variable)
X = filtered_data_no_outliers["TOT_POWER_MW"].values.reshape(-1, 1)
y = filtered_data_no_outliers["TOT_DATACENTER_SPACE_SQFT"].values

### Exponential Regression ###
# Transform y to log scale
y_log = np.log(y)

# Fit a linear model to the log-transformed data
exp_model = LinearRegression()
exp_model.fit(X, y_log)
y_pred_exp = np.exp(exp_model.predict(X))  # Transform back to original scale

# Calculate MSE and R^2 for exponential regression
mse_exp = mean_squared_error(y, y_pred_exp)
r2_exp = r2_score(y, y_pred_exp)

### Polynomial Regression (Degree 2) ###
# Transform the data to include polynomial terms
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit a linear model to the polynomial features
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_pred_poly = poly_model.predict(X_poly)

# Calculate MSE and R^2 for polynomial regression
mse_poly = mean_squared_error(y, y_pred_poly)
r2_poly = r2_score(y, y_pred_poly)

# Print metrics for both models
print("Exponential Regression:")
print(f"Mean Squared Error (MSE): {mse_exp:.2f}")
print(f"R^2 Value: {r2_exp:.2f}\n")

print("Polynomial Regression (Degree 2):")
print(f"Mean Squared Error (MSE): {mse_poly:.2f}")
print(f"R^2 Value: {r2_poly:.2f}")

# Plot the scatter plot and best-fit lines
plt.figure(figsize=(12, 8))
plt.scatter(X, y, alpha=0.7, label="Actual Data")

# Generate a smooth range of X values for plotting the best fit
X_fit = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)

# Plot exponential regression best-fit line
y_fit_exp = np.exp(exp_model.predict(X_fit))
plt.plot(X_fit, y_fit_exp, color="orange", label="Exponential Fit", linewidth=2)

# Plot polynomial regression best-fit line
X_fit_poly = poly.fit_transform(X_fit)
y_fit_poly = poly_model.predict(X_fit_poly)
plt.plot(X_fit, y_fit_poly, color="green", label="Polynomial Fit (Degree 2)", linewidth=2)

plt.title("Exponential and Polynomial Regression")
plt.xlabel("Total Power Usage (MW)")
plt.ylabel("Total Data Center Space (sq ft)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'lst_us_data_centers_20241115_070225.csv'  # Update the file path as needed
data = pd.read_csv(file_path)

# Filter rows where both "TOT_DATACENTER_SPACE_SQFT" and "TOT_POWER_MW" have values
filtered_data = data.dropna(subset=["TOT_DATACENTER_SPACE_SQFT", "TOT_POWER_MW"])

# Convert columns to numeric
filtered_data["TOT_DATACENTER_SPACE_SQFT"] = pd.to_numeric(filtered_data["TOT_DATACENTER_SPACE_SQFT"], errors='coerce')
filtered_data["TOT_POWER_MW"] = pd.to_numeric(filtered_data["TOT_POWER_MW"], errors='coerce')

# Drop rows with invalid numeric data
filtered_data = filtered_data.dropna(subset=["TOT_DATACENTER_SPACE_SQFT", "TOT_POWER_MW"])

# Apply Z-score method to remove outliers
z_scores = np.abs(zscore(filtered_data[["TOT_POWER_MW", "TOT_DATACENTER_SPACE_SQFT"]]))
filtered_data_no_outliers = filtered_data[(z_scores < 2).all(axis=1)]
filtered_data_no_outliers = filtered_data_no_outliers[
    ~((filtered_data_no_outliers["TOT_POWER_MW"] * 1000 == 100000) |
      (filtered_data_no_outliers["TOT_POWER_MW"] * 1000 == 105000) | 
      (filtered_data_no_outliers["TOT_POWER_MW"] * 1000 == 84000))
]

# Define X (independent variable) and y (dependent variable)
X = filtered_data_no_outliers["TOT_POWER_MW"].values.reshape(-1, 1)
y = filtered_data_no_outliers["TOT_DATACENTER_SPACE_SQFT"].values

# Multiply all X values by 1000
X = X * 1000

### Linear Regression ###
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

# Calculate metrics for linear regression
mse_linear = mean_squared_error(y, y_pred_linear)
r2_linear = r2_score(y, y_pred_linear)

### Exponential Regression ###
# Transform y to log scale
y_log = np.log(y)
exp_model = LinearRegression()
exp_model.fit(X, y_log)
y_pred_exp = np.exp(exp_model.predict(X))

# Calculate metrics for exponential regression
mse_exp = mean_squared_error(y, y_pred_exp)
r2_exp = r2_score(y, y_pred_exp)

### Logarithmic Regression ###
# Transform X to log scale
X_log = np.log(X)
log_model = LinearRegression()
log_model.fit(X_log, y)
y_pred_log = log_model.predict(X_log)

# Calculate metrics for logarithmic regression
mse_log = mean_squared_error(y, y_pred_log)
r2_log = r2_score(y, y_pred_log)

# Print metrics for all models
print("Linear Regression:")
print(f"  MSE: {mse_linear:.2f}, R^2: {r2_linear:.2f}")
print("\nExponential Regression:")
print(f"  MSE: {mse_exp:.2f}, R^2: {r2_exp:.2f}")
print("\nLogarithmic Regression:")
print(f"  MSE: {mse_log:.2f}, R^2: {r2_log:.2f}")

# Plot the data and all regression fits
plt.figure(figsize=(14, 8))
plt.scatter(X, y, color="blue", alpha=0.6, label="Actual Data")

# Generate a smooth X range for plotting best-fit curves
X_fit = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)

# Linear Fit
y_fit_linear = linear_model.predict(X_fit)
plt.plot(X_fit, y_fit_linear, color="red", label="Linear Fit", linewidth=2)

# Exponential Fit
y_fit_exp = np.exp(exp_model.predict(X_fit))
plt.plot(X_fit, y_fit_exp, color="orange", label="Exponential Fit", linewidth=2)

# Logarithmic Fit
X_fit_log = np.log(X_fit)
y_fit_log = log_model.predict(X_fit_log)
plt.plot(X_fit, y_fit_log, color="green", label="Logarithmic Fit", linewidth=2)

plt.title("Linear, Exponential, and Logarithmic Regression (Best-Fit Curves)")
plt.xlabel("Total Power Usage (kW)")
plt.ylabel("Total Data Center Space (sq ft)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Combine X and y into a single DataFrame for easier manipulation
data = pd.DataFrame({"X": X.flatten(), "Y": y})

# Group by X and compute the mean of Y values for duplicate X
averaged_data = data.groupby("X", as_index=False).mean()

# Extract the averaged X and Y values
X_avg = averaged_data["X"].values.reshape(-1, 1)
y_avg = averaged_data["Y"].values

# Polynomial regression
poly_model = make_pipeline(PolynomialFeatures(degree=3), LinearRegression())
poly_model.fit(X_avg, y_avg)
y_fit_poly = poly_model.predict(X_avg)

# Linear regression
linear_model = LinearRegression()
linear_model.fit(X_avg, y_avg)
y_fit_linear = linear_model.predict(X_avg)

# Logarithmic regression (fit with log-transformed X)
log_X_avg = np.log(X_avg + 1e-8)  # Add a small value to avoid log(0)
log_model = LinearRegression()
log_model.fit(log_X_avg, y_avg)
y_fit_log = log_model.predict(log_X_avg)

# Plotting
plt.figure(figsize=(14, 8))

# Scatter plot of the averaged data
plt.scatter(X_avg, y_avg, alpha=0.7, label="Averaged Data", color="blue")

# Polynomial fit
X_fit = np.linspace(X_avg.min(), X_avg.max(), 500).reshape(-1, 1)
y_fit_poly_extended = poly_model.predict(X_fit)
plt.plot(X_fit, y_fit_poly_extended, color="green", label="Polynomial Regression (Degree 3)", linewidth=2)

# Linear fit
y_fit_linear_extended = linear_model.predict(X_fit)
plt.plot(X_fit, y_fit_linear_extended, color="red", label="Linear Regression", linewidth=2)

# Logarithmic fit
log_X_fit = np.log(X_fit + 1e-8)
y_fit_log_extended = log_model.predict(log_X_fit)
plt.plot(X_fit, y_fit_log_extended, color="purple", label="Logarithmic Regression", linewidth=2)

# Labels and Legend
plt.title("Best-Fit Lines: Polynomial, Linear, and Logarithmic")
plt.xlabel("Total Power Usage (kW)")
plt.ylabel("Total Data Center Space (sq ft)")
plt.legend()
plt.savefig("landuseRegression")
plt.show()

# Print Metrics
print("Metrics:")
print(f"Polynomial Regression R^2: {r2_score(y_avg, y_fit_poly):.2f}")
print(f"Linear Regression R^2: {r2_score(y_avg, y_fit_linear):.2f}")
print(f"Logarithmic Regression R^2: {r2_score(y_avg, y_fit_log):.2f}")
```
